# **⭐ Classification Model to Identify Multiple Disease**

```python
# import library
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# define y
y=df['prognosis']
# define X
x=df.drop(['prognosis'],axis=1)
```

```python
# split data
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,random_state=21,test_size=0.30)
```

```python
# verify shape
xtrain.shape
xtest.shape
```

```python
# select model
from sklearn.naive_bayes import GaussianNB
```

```python
# train model
gauss=GaussianNB()
model=gauss.fit(xtrain,ytrain)
```

```python
# predict with model
ypred=model.predict(xtest)
ypred
```

```python
# model accuracy
from sklearn.metrics import accuracy_score
accuracy_score(ytest,ypred)
```

```python
# model confusion matrix
from sklearn.metrics import confusion_matrix
confusion_matrix(ytest,ypred)
```

```python
# future prediction
# define X_new
newx=xtest.sample()
newx
# predict for X_new
pre_newx=model.predict(newx)
pre_newx
```
